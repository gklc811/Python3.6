```python
import matplotlib.pyplot as plt
```

```python
a=[1,2,3]
b=[1,4,6]
plt.plot(a,b)
plt.ylabel("y axis")
plt.xlabel("x axis")
plt.title("Sample")
plt.grid()
```

```python
a=[1,2,3]
b=[1,4,6]
plt.plot(a,b,'go')
plt.ylabel("y axis")
plt.xlabel("x axis")
plt.title("Sample")
plt.grid()
```

```python
a=[1,2,3]
b=[1,4,6]
plt.plot(a,b,'g*')
plt.ylabel("y axis")
plt.xlabel("x axis")
plt.title("Sample")
plt.grid()
```

```python
import numpy as np
a = np.arange(-5.,5.2,.2)

plt.plot(a, a, 'b*', label = "straight")
plt.plot(a, a**2, 'g--', label = "square")
plt.plot(a, a**3, 'r^', label = "cube")
plt.legend()
plt.grid()
plt.show()
```

```python
a = np.arange(-5.,5.2,.2)
plt.plot(a, a,  linewidth = 6.0, color = 'r')
plt.show()
```
